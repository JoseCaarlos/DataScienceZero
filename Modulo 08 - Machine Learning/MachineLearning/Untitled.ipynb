# Teste de Overfittinng

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
```

```python
dataset = pd.read_csv('../SVM/Materiais de Apoio/data.csv', sep=',')
```

```python
dataset.head()
```

```python
enc = LabelEncoder()
```

```python
inteiro = enc.fit_transform(dataset['artist'])
```

```python
set(inteiro)
```

```python
dataset['artist'] = inteiro
dataset.drop(labels=['song_title', 'id'], axis=1, inplace=True)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
# Utiliza a função train_test_split para separar os dados de treino e teste em 80/20
# Random_state -> Escolha forma aleatório de escolha dos dados
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset.drop('target', axis=1), dataset['target'], test_size=0.2, random_state=123)
```

```python
y_train
```

```python
# Scala os dados de treino e teste
scaler = StandardScaler().fit(X_train)
x_train = scaler.transform(X_train)

scaler2 = StandardScaler().fit(X_test)
x_test = scaler2.transform(X_test)
```

```python
X_test
```

```python
# Treina o classificado SVM
clf.fit(X_train, y_train)
```

```python
# Resultado da predição
y_pred = clf.predict(X_test)
```

```python
# Imprime a acurácia do modelo
metrics.accuracy_score(y_test,y_pred)
```
